# **What is Logistic Regression**
Logistic Regression is a type of regression that is used for predicting (classifying) categorical values. Unlike linear regression, which predicts continuous values, logistic regression predicts categorical variables (Mehta, 2019).

There are two types of logistic regression: Binary logistic regression and multnomial logistic regression
- Binary logistic regression involves the classification of two categories
- Multinomial logistic regression involves the classification of three or more categories

Binary logistic Regression learns by learning the decision boundary between the two classes. Multnomial logistic reression calculates the probability that a certain data point belongs into one of the classes. The class with the highest probability is considered the class in which that data point belongs to.

### The Process Behind Logistic Regression
Logistic regression takes in features as inputs and outputs a target class.
- First, logistic converts inputs into probabilities, using activation fuctions, such as sigmoid or softmax.
- However, these predicted probabilities may differ from the actual probabilities. The difference between the predicted probabbilities and actual probabilities is called a cost/loss function.
- The loss function needs to be minimized in order to increase model accuracy. To minimize the error (loss function), the model undergoes a gradient descent process, which iteratively updates the weights and biases (Mehta, 2019).
-The process continues until the probabilities converge. The probabilities will then be converted into the target categories.


### Binary Logistic Regression

- Binary Logistic regression consists of a sigmoid function and a cost function (Mehta, 2019)
- Inputs are entered into the model
- A weighted sum is calculated using the weights and biases
- The weighted sum is passed through the sigmoid function which produces the predicted probabbility
- The cost/loss function calculates the error between the predicted probability and actual probability
- Gradient descent is used to minimize the cost function by updating the weights (Mehta, 2019)
- The process is repeated until the model convereges
- It uses thresholding to convert probabilities into target classes. If the predicted probability => 0.5, the input is classified into the positive class (1); if the probability < 0.5, it's classified into the negative class (0).

### Multinomial Logistic Regression

- Inputs are entered into the model
- The scores are calculated for each class
- The softmax function transforms the scores into probabilities that each data points belongs to a class (Mahmoud,2018). It generates a probability distribution across all classes. The class with the highest probability is typically chosen as the final predicted class for the input
- The cost function calculates the error between the predicted
- Weights and biases are updated using gradient descent (Mehta, 2019)
- Using the updated weights, the process is repeated until the model converges
- It uses ArgMax to convert probabilities into output classes. The class with the highest probability is selected as the predicted class for the input.

# **Model Evaluation & Improvement**
## How the Model was Evaluated

1. **Acuracy scores**
- The training and test scores were able to suggest whether my model was overfitting or nor. If the training accuracy is significantly larger than the test accuracy, this means the model is overfitting. If both the test and training scores are low, it suggests an underfitting model
- In this case, the model was found to be underfitting
2. **Cross validation**
  - Crossvalidation helped me check if the model is overfitting or underfitting. A low cross validation average usually means that the model is underfitting. In this case, the model was confirmed to be underfitting.

3. **Precision:**
   - Precision helped me determine the proportion of true positive predictions among all the reviews that have been predicted as belonging to a certain rating.
   - It's given by Precision = True Positives / (True Positives + False Positives) (Agrawal, 2021)

4. **Recall:**
   - Recall helped me calculate the proportion of true positive predictions among all actual positives. Recall helped me understand among the reviews classified into a certain category, how many reviews actualy belong to that category.
   - It's given by Recall = True Positives / (True Positives + False Negatives)

5. **F1 Score:**
   - The F1 score helped me get a balanced evaluation of the model's performance. This is because it is a balance between precision and recall.
   - F1 Score = 2 * (Precision * Recall) / (Precision + Recall) (Agrawal, 2021)

6. **Confusion Matrix**
  - The confusion matrix is useful for understanding where the model is making errors and the types of errors being made.

6. **Loss/Accuracy Plot:**
   - The Loss and Accuracy plots helped me visualize the model's learning progress, helping me identify issues like overfitting or underfitting.
  

7. **ROC Curve:**
   - A ROC curve helped me visualize my model's ability to distinguish between multiple classes. It plots the true positive rate against the false positive rate, helping me understand the extent of each class's performance relating to true positives and false positives (Zach, 2021).

# **Creating Logistic Regression Class**
- Creating a class called 'LogReg'

```python
import numpy as np
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import KFold
import seaborn as sns
from sklearn.model_selection import cross_val_score

class LogReg:

#1.Initializing Weights and Biases (Learning Rate and Epochs)
    def __init__(self, learning_rate=0.01, num_iterations=1000):
        self.learning_rate = learning_rate
        self.num_iterations = num_iterations
        self.weights = None
        self.biases = None

#Initializing the weights and biases
# Setting biases to 0 and weights to random
    def initialize_weights(self, n_features, n_classes):
        self.weights = np.random.rand(n_features, n_classes)
        self.biases = np.zeros(n_classes)

# 2. Defining a softmax function, using the Softmax formula:
    def softmax_func(self, z):
        exp_z = np.exp(z - np.max(z, axis=1, keepdims=True))
        return exp_z / np.sum(exp_z, axis=1, keepdims=True)

#3. Defining the loss function, using the formula for categorical crossentropy
    def loss_func(self, y, yp):
        epsilon = 1e-15
        yp = np.clip(yp, epsilon, 1 - epsilon) #clip () used to avoid extreme values
        loss = -np.log(yp[range(len(y)), y]) # Extracting the predicted probabilities for the correct classes
        return np.mean(loss)

# Calculating gradient descent using the formula for gradient descent
    def grad_desc(self, X, y, n_classes):
        m = len(y)
        n_features = X.shape[1]
        self.initialize_weights(n_features, n_classes)

        for i in range(self.num_iterations):
            z = np.dot(X, self.weights) + self.biases #calculate logit score
            yp = self.softmax_func(z) #applying softmax to get the probabilities

            one_hot_y = np.zeros((m, n_classes)) #perfroming one hot encoding for target labels
            one_hot_y[range(m), y] = 1

            dz = yp - one_hot_y
            dw = (1 / m) * np.dot(X.T, dz) #gradient descent formula for weights
            db = (1 / m) * np.sum(dz, axis=0) #gradient descent for biases

            self.weights -= self.learning_rate * dw
            self.biases -= self.learning_rate * db

#Creating a function to fit the data:
    def fit(self, X, y, **kwargs):
        n_classes = np.max(y) + 1  # calculating number of classes
        self.grad_desc(X, y, n_classes) #

#Creating a function to predict the data:
    def predict(self, X):
        z = np.dot(X, self.weights) + self.biases #Linear tranformation of wighted sum + bias
        yp = self.softmax_func(z) #applying softmax function
        return np.argmax(yp, axis=1) #returning class labels - converting probablities to labels using ArgMax

# Predicting probability of classes in multinomial logistic regression using predict_proba
    def predict_proba(self, X):
        z = np.dot(X, self.weights) + self.biases
        yp = self.softmax_func(z)
        return yp #returing the probabilities

#Creating a function to normalize the data
    def normalize(self, X):
      m,n=X.shape
      for i in range(n):
        X=(X-X.mean(axis=0))/X.std(axis=0)
      return X

#Creating a function to evaluate the model
    def evaluate(self, X, y):
        y_pred = self.predict(X) #predicting isng predict() function
        accuracy = accuracy_score(y, y_pred) #calculating accuracy score
        report = classification_report(y, y_pred) #classification report
        conf_matrix = confusion_matrix(y, y_pred) #confusion matrix
        #conf_heat = sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues") #creating heap map from confusion matrix
        return accuracy, report, conf_matrix #returning the accuracy score class. report and confusion matrix

# Setting params for Crossvalidation
    def set_params(self, **parameters):
        for parameter, value in parameters.items():
            setattr(self, parameter, value)
            return self
    #Code adapted from: https://stackoverflow.com/questions/28124366/can-gridsearchcv-be-used-with-a-custom-classifier
        #https://stackoverflow.com/questions/20330445/how-to-write-a-custom-estimator-in-sklearn-and-use-cross-validation-on-it

# Getting params for crossvalidation
    def get_params(self, deep=False):
        return {'learning_rate': self.learning_rate, 'num_iterations': self.num_iterations}
   #Code adapted from: https://stackoverflow.com/questions/28124366/can-gridsearchcv-be-used-with-a-custom-classifier

#Creating a function for hyperparameter tuning
    def hyperparameter_tuning(self, X, y, learning_rates=[0.01, 0.1, 0.5], num_iterations_list=[1000, 2000, 3000]):
        best_accuracy = 0.0
        best_lr = None
        best_iterations = None

        for lr in learning_rates:
            for iterations in num_iterations_list: # Iterate over different numbers of iterations
                self.learning_rate = lr
                self.num_iterations = iterations

                mean_accuracy = self.cross_validate(X, y)

                if mean_accuracy > best_accuracy: # If the accuracy is better than the best accuracy found so far
                    best_accuracy = mean_accuracy # Update the best accuracy
                    best_lr = lr # Update the best learning rate
                    best_iterations = iterations # Update the best number of iterations

        return best_lr, best_iterations, best_accuracy # Return the best hyperparameters found
#Code adapted from: https://www.kaggle.com/code/vitorgamalemos/multinomial-logistic-regression-from-scratch
```

##Tokenizing

Tokenization is the breaking up of a phrase, sentence or paragraph into smaller units such as individual words (called tokens), that a program can easily work with. Without tokenization, it's impossible to carry out analysis of text data.

Keras Tokenizer is used to tokenize the data (saturncloud.io, 2023).
- Keras tokenizer assigns each word a unique integer id after tokenization.
- The Tokenizer converts the text sequences into sequences of integers using the vocabulary. Each word in the text is replaced with its integer ID, creating a sequence of integers representing the original text (saturncloud.io, 2023).
- The Tokenizer can also pad or truncate the sequences to a fixed length, ensuring that all sequences have the same length. This is important for training LSTMs, as inputs must have consistent dimensions.

- The Tokenizer provides options to handle out-of-vocabulary words that are not present in the vocabulary. An approach is to replace them with a special "out-of-vocabulary" token.

Thus, the tokenizer makes use of the following parameters

1. **'max_length' (Max Sequence Length):**
   - 'max_length' is the maximum length of a sequence  after tokenization and padding. Sequences longer than this will be truncated, and sequences shorter than this will be padded.

2. **'trunc type' (Truncation Type):**
   - It specifies whether truncation should be done at the beginning or end of the sequence if it exceeds the 'max_length'. In this analysis, 'post' indicates truncation at the end of the sequence.

3. **'padding_type' (Padding Type):**
   - It determines whether padding should be added at the beginning or end of the sequence. In this analysis, 'post' indicates padding at the end of the sequence.

4. **'oov_token' (Out-of-Vocabulary Token):**
   - The out-of-vocabulary token is a special token used to represent words that are not present in the vocabulary. It is used during tokenization to handle out-of-vocabulary words.

5. **'vocab' (Vocabulary Size):**
   - 'vocab' specifies the desired vocabulary size, that is, the number of unique words to include in the vocabulary.



The following values were chosen for the parameters:

```python
#tokenize
max_length = 200 #the max_length was chosen as 200
trunc_type = 'post'
padding_type = 'post'
oov_token = '<PAD>'
vocab = 700 #the vocabulary size was chosen as 700
```

The classification report uses the results of the confusion matrix.
- Precision is the ratio of the number of samples actually belonging to the positive class and all the samples that were predicted to be of the positive class.
Precision= True Positives/(False Positives+True Positives)

- Recall is the ratio of the number of samples predicted correctly to be belonging to the positive class and all the samples that actually belong to the positive class.
Recall = True Positives/(False Negatives+True Positives)
- f1 Score is the harmonic mean of the recall and precision scores from the positive class. The f1-score takes into account both the precision and recall, and a high f1-score (closer to 1) implies that both the precision and recall are high.

#Plot

I will be creating an ROC curve to analyze the data.

The Receiver Operating Characteristic (ROC) curve is used for evaluating the performance of a classification model.

The plot shows the relationship between the true positive rate (TPR) and the false positive rate (FPR).

- A high TPR indicates that the model is good at identifying true positives.

- A low FPR indicates that the model is good at not falsely labeling negatives as positives.

The ROC curve is plotted with TPR on the y-axis and FPR on the x-axis. An ideal ROC curve is one that hugs the top-left corner of the plot, because this would indicate high TPR and low FPR.

Along with the ROC curve, the area under the ROC curve (AUC) is a  metric used to determine the model's performance. AUC outputs a single numerical value, which represents the model's ability to discriminate between positive and negative samples. The AUC is measured between 0 and 1. A higher AUC (closer to 1) indicates better performance.

The ROC is typically used for evaluating the performance of a binary classification model. However, in this analysis it was used to evaluate the performance of a multiclass classification model.

#Boosting

- Attempting to improve the model by applying boosting to the model
- Boosting is a ensemble learning technique that combines multiple weak learners to form a strong learner (Kalirane, 2023)
- In this case, the logreg model is a weak learner because it underfits (it has a high bias)
- Boosting combines models with a high bias to form one with a reduced bias
- The aim behind generalization is to have a low bias and a low variance. Thus applying boosting to the model will likely reduce the bias.

I first will create a class BoostLogReg, similar to the LogReg class:

```python
import numpy as np
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import KFold
import seaborn as sns
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import AdaBoostClassifier

class BoostLogReg:

#Step 1: Initialize Weights and Biases (Learning Rate and Epochs)
    def __init__(self, learning_rate=0.01, num_iterations=1000):
        self.learning_rate = learning_rate
        self.num_iterations = num_iterations
        self.weights = None
        self.biases = None

#Initializing the weights and biases
# Setting biases to 0 and weights to random
    def initialize_weights(self, n_features, n_classes):
        self.weights = np.random.rand(n_features, n_classes)
        self.biases = np.zeros(n_classes)

# Defining a softmax function, using the Softmax formula:
    def softmax_func(self, z):
        exp_z = np.exp(z - np.max(z, axis=1, keepdims=True))
        return exp_z / np.sum(exp_z, axis=1, keepdims=True)

#Defining the loss function, using the formula for categorical crossentropy
    def loss_func(self, y, yp):
        epsilon = 1e-15
        yp = np.clip(yp, epsilon, 1 - epsilon) #clip () used to avoid extreme values
        loss = -np.log(yp[range(len(y)), y]) # Extracting the predicted probabilities for the correct classes
        return np.mean(loss)

# Calculating gradient descent using the formula for gradient descent
    def grad_desc(self, X, y, n_classes):
        m = len(y)
        n_features = X.shape[1]
        self.initialize_weights(n_features, n_classes)

        for i in range(self.num_iterations):
            z = np.dot(X, self.weights) + self.biases
            yp = self.softmax_func(z)

            one_hot_y = np.zeros((m, n_classes))
            one_hot_y[range(m), y] = 1

            dz = yp - one_hot_y
            dw = (1 / m) * np.dot(X.T, dz)
            db = (1 / m) * np.sum(dz, axis=0)

            self.weights -= self.learning_rate * dw
            self.biases -= self.learning_rate * db

# #Creating a function to fit the data:

    def fit(self, X, y, sample_weight=None):
        num_samples, num_features = X.shape
        self.num_classes = np.max(y) + 1  #  the number of classes
        self.classes_ = np.unique(y)  # the unique class labels
        self.grad_desc(X, y, self.num_classes)  # Calling gradient descent for training

#Creating a function to predict the data:
    def predict(self, X):
        z = np.dot(X, self.weights) + self.biases #Linear tranformation of wighted sum + bias
        yp = self.softmax_func(z) #applying softmax function
        return np.argmax(yp, axis=1) #returning class labels - converting probablities to labels using ArgMax


# Predicting probability of classes in multinomial logistic regression using predict_proba
    def predict_proba(self, X):
        z = np.dot(X, self.weights) + self.biases
        yp = self.softmax_func(z) #applying the sofmax func
        return yp #returing the probabilities

#Creating a function to normalize the data
    def normalize(self, X):
      m,n=X.shape
      for i in range(n):
        X=(X-X.mean(axis=0))/X.std(axis=0)
      return X

#Creating a function to evaluate the model
    def evaluate(self, X, y):
        y_pred = self.predict(X) #predicting isng predict() function
        accuracy = accuracy_score(y, y_pred) #calculating accuracy score
        report = classification_report(y, y_pred) #classification report
        conf_matrix = confusion_matrix(y, y_pred) #confusion matrix
        #conf_heat = sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues") #creating heap map from confusion matrix
        return accuracy, report, conf_matrix #returning the accuracy score class. report and confusion matrix

# Setting params for Crossvalidation
    def set_params(self, **parameters):
        for parameter, value in parameters.items():
            setattr(self, parameter, value)
            return self
    #Code adapted from: https://stackoverflow.com/questions/28124366/can-gridsearchcv-be-used-with-a-custom-classifier
        #https://stackoverflow.com/questions/20330445/how-to-write-a-custom-estimator-in-sklearn-and-use-cross-validation-on-it

# Getting params for crossvalidation
    def get_params(self, deep=False):
        return {'learning_rate': self.learning_rate, 'num_iterations': self.num_iterations}
   #Code adapted from: https://stackoverflow.com/questions/28124366/can-gridsearchcv-be-used-with-a-custom-classifier

#Creating a function for hyperparameter tuning
    def hyperparameter_tuning(self, X, y, learning_rates=[0.01, 0.1, 0.5], num_iterations_list=[1000, 2000, 3000]):
        best_accuracy = 0.0
        best_lr = None
        best_iterations = None

        for lr in learning_rates:
            for iterations in num_iterations_list: # Iterate over different numbers of iterations
                self.learning_rate = lr
                self.num_iterations = iterations

                mean_accuracy = self.cross_validate(X, y)

                if mean_accuracy > best_accuracy: # If the accuracy is better than the best accuracy found so far
                    best_accuracy = mean_accuracy # Update the best accuracy
                    best_lr = lr # Update the best learning rate
                    best_iterations = iterations # Update the best number of iterations

        return best_lr, best_iterations, best_accuracy # Return the best hyperparameters found
#Code adapted from: https://www.kaggle.com/code/vitorgamalemos/multinomial-logistic-regression-from-scratch
```
