# What is a Recurrent Neural Network
A Recurrent Neural Network (RNN) is a type of neural network that is able to work with sequential data. Sequential data includes data where the current element depends on the previous element, such as time series data or words in sentences. RNNs are able to work with sequential data, because they contain mechanisms that allow keep memory in a sequential manner.  Unlike other types of neural networks, RNNs contain loops that enable them to keep memory from previous input (Starmer, 2022). Keeping memory allows them to understand the context of previous input, allowing them to predict future values.

RNNs work in the following way:
- RNNs typically consist of an input layer, hidden layer(s) and an output layer
-	An input layer - RNNs take in sequential data, which is taken in at each time step and passed to the next layer.
-	Recurrent layer - the recurrent layer, also known as the hidden layer contains what allows the RNN to keep memory: the hidden state.  Then hidden state represents the RNNs stored memory at a certain time step. The hidden state is passed to the next time step (Talaviya, 2022).
-	The output layer is the final layer of the RNN. It processes the hidden state at each time step and produces output for that time step.  This layer also contains the activation function. The activation function introduces non-linearity to the model (for classification models), allowing the model to learn complex patterns in data, and allowing it to properly approximate real world data. Depending on the task, such as classification or regression, the activation function may vary (El Boujnouni, 2021; Ayachi, 2021).

In order for RNNs to make accurate predictions, their loss function (error between actual and predicted values) need to be minimized. In order to minimize the loss function, RNNs are trained using a process called backpropagation. Backpropagation aims to find the optimal combination of weights (Rouse, 2023) in order for minimizing the loss function. The weights refer to the parameters learnt from the input data. Backpropagation works in a iterative manner whereby the model's outputs are propagated back through the model in order to readjust the weights and produce a more accurate output (Rouse, 2023).

The disadvantage of the RNN is that it is prone to vanishing and exploding gradient problems (Starmer, 2022). These are caused when the gradients of the weights become extremely large or small when propagated through the network during training (Starmer, 2022). This can prevent the model from learning effectively and converging. During backpropagation, gradients are propagated via recurrent connections which can cause large gradients to accumulate and small gradients to become infinitely small (Starmer, 2022).

Long Short-Term Memory (LSTM) is a type of RNN that is able to address the vanishing and exploding gradient problems (Starmer, 2022). This is because they are able to capture long term dependencies in sequential data.

The architecture of an LSTM network consists of four components.These are a cell, an input gate, a forget gate, and an output gate (Singh, 2021).
- The cell's function is to store and update the memory state
- The input gate's function is to determine the amount of new information stored in the memory
- The forget gate's function is to decide the information discarded from the memory
- The output gate's function is to regulate  the amount of information outputted from the cell (Singh, 2021).

The architecture of the LSTM is perfect for capturing long-term dependencies because it enables the LSTM to remember or forget information across long sequences.

#### **The Steps Taken to Evaluate the Model**
In order to evaluate my model, I used the following measures:
1. **Loss and Accuracy**
   - **Loss:** The loss helped me measure of how well my model performed during training. This is because it measures the difference between the predicted values and the actual values. A lower loss value would mean a better fit of the model to the training data.
   - **Accuracy:** The accuracy helped me measure the total number of correctly classified reviews. This is because it measures the proportion of correctly classified instances out of the total instances. Higher accuracy indicates a better-performing model overall.

2. **Confusion Matrix:**
   - The confusion matrix helped me get a deeper understanding of my model's performance, by giving me the true extent of its accuracy. To do this, it displayed the counts of true positives, true negatives, false positives, and false negatives (Agrawal, 2021)
   - True Positives TP: Showed the number of reviews that the model correctly predicted as belonging to a certain star rating
   - False Positives FP: Showed the number of reviews that the model incorrectly predicted as belonging to a certain star rating when it does not
   - False Negatives FN: Showed the number of reviews that the model incorrectly predicted as not belonging to a certain star rating when it does
   - True Negatives TN: Showed the number of reviews that the model correctly predicted as not belonging to a certain star rating

The confusion matrix is useful for understanding where the model is making errors and the types of errors being made.

3. **Precision:**
   - Precision helped me determine the proportion of true positive predictions among all the reviews that have been predicted as belonging to a certain rating.
   - It's given by Precision = True Positives / (True Positives + False Positives) (Agrawal, 2021)

4. **Recall:**
   - Recall helped me calculate the proportion of true positive predictions among all actual positives. Recall helped me understand among the reviews classified into a certain category, how many reviews actualy belong to that category.
   - It's given by Recall = True Positives / (True Positives + False Negatives)

5. **F1 Score:**
   - The F1 score helped me get a balanced evaluation of the model's performance. This is because it is a balance between precision and recall.
   - F1 Score = 2 * (Precision * Recall) / (Precision + Recall) (Agrawal, 2021)

6. **Loss/Accuracy Plot:**
   - The Loss and Accuracy plots helped me visualize the model's learning progress, helping me identify issues like overfitting or underfitting.

7. **ROC Curve:**
   - A ROC curve helped me visualize my model's ability to distinguish between multiple classes. It plots the true positive rate against the false positive rate, helping me understand the extent of each class's performance relating to true positives and false positives (Zach, 2021).

#### **Dealing with Overfitting**
As previously mentioned, to deal with overfitting, I used dropout regularization. Dropout is a regularization technique used to reduce overfitting in neural networks.

In dropout regularization, during training, a part of the neurons in a layer are randomly "dropped out" (Brownlee, 2018). This means that those neurons' contributions to the forward pass and backpropagation are temporarily removed. The dropout rate is the percentage of the neurons that are dropped out.

By randomly dropping out neurons, the network is forced to learn other features, because it can't rely heavily on any specific set of neurons (SEb, 2021). This often results in a more generalized model that performs better on unseen data.

In my case, I used hyperparameter tuning to help me decide the optimal dropout rate.

## Tokenize Data
Tokenization is the breaking up of a phrase, sentence or paragraph into smaller units such as individual words (called tokens), that a program can easily work with. Without tokenization, it's impossible to carry out analysis of text data.

Keras Tokenizer is used to tokenize the data (saturncloud.io, 2023).
- Keras tokenizer assigns each word a unique integer id after tokenization.
- The Tokenizer converts the text sequences into sequences of integers using the vocabulary. Each word in the text is replaced with its integer ID, creating a sequence of integers representing the original text (saturncloud.io, 2023).
- The Tokenizer can also pad or truncate the sequences to a fixed length, ensuring that all sequences have the same length. This is important for training LSTMs, as inputs must have consistent dimensions.

- The Tokenizer provides options to handle out-of-vocabulary words that are not present in the vocabulary. An approach is to replace them with a special "out-of-vocabulary" token.

### **Tokenizer Parameters**

Thus, the tokenizer makes use of the following parameters

1. **'max_length' (Max Sequence Length):**
   - 'max_length' is the maximum length of a sequence  after tokenization and padding. Sequences longer than this will be truncated, and sequences shorter than this will be padded.

2. **'trunc type' (Truncation Type):**
   - It specifies whether truncation should be done at the beginning or end of the sequence if it exceeds the 'max_length'. In this analysis, 'post' indicates truncation at the end of the sequence.

3. **'padding_type' (Padding Type):**
   - It determines whether padding should be added at the beginning or end of the sequence. In this analysis, 'post' indicates padding at the end of the sequence.

4. **'oov_token' (Out-of-Vocabulary Token):**
   - The out-of-vocabulary token is a special token used to represent words that are not present in the vocabulary. It is used during tokenization to handle out-of-vocabulary words.

5. **'vocab' (Vocabulary Size):**
   - 'vocab' specifies the desired vocabulary size, that is, the number of unique words to include in the vocabulary.



The following values were chosen for these parameters:

```python
max_length = 200 #the max_length was chosen as 200
trunc_type = 'post'
padding_type = 'post'
oov_token = '<PAD>'
vocab = 700 #the vocabulary size was chosen as 700
```

# Model Building

Initializing the parameters required in the LSTM model.
- 'embedding dim' (embedding dimension) determines the size of the vector to represent each word.
- drop_value is the dropout rate, which is the probability chosen for dropout regularization. The chosen rate is 0.2.
- n_dense is the number of dense or fully connected layers in the network. The number of layers chosen in this analysis was 5. This is because this is a multiclass classification problem and the number of fully connected layers must correspond to the number of classes, which is 5.

```python
#parameters
vocab_size = 1000
embedding_dim = 16
drop_value = 0.2
n_dense = 5
```

Building an LSTM model:
The architecture of an LSTM model is as follows:
- ***Sequential()*** is used to define a sequential model, eg. LSTM.

- **model.add(Embedding(vocab_size, embedding_dim, input_length=max_length))** adds an embedding layer. This layer is responsible for word embedding, where words are represented as dense vectors.
   - 'input_length' is the length of the input sequences.

- **model.add(LSTM(30))** adds an LSTM layer with 30 nodes.

- **model.add(Dropout(drop_value))** adds a dropout layer with a specified probability (drop_value) to prevent overfitting.

- **model.add(Dense(n_dense, activation='softmax'))** adds a dense layer with 'n_dense' nodes and a softmax activation function. This is the output layer for multi-class classification.

```python
# Build the LSTM model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, GlobalAveragePooling1D, Dense, Dropout
model = Sequential()
model.add(Embedding(vocab_size, embedding_dim, input_length=max_length))
model.add(LSTM(30))
model.add(Dropout(drop_value))
model.add(Dense(n_dense, activation='softmax'))  # Output layer with 2 units (binary classification)
import tensorflow as tf
tf.random.set_seed(42)

# Display the model summary
model.summary()
```

The optimizer used for gradient descent in this analysis is the Adam optimizer.  What Adam Optimizer does is it helps specify the method for updating the model's weights (Geeksforgeeks, n.d.)

## Classification Report

I will calculate the Classification report. The advantage of the classification report is that it provides a summary of the model's performance across different classes, according to the metrics of precision, recall, and F1-score.

The classification report uses the results of the confusion matrix.
- Precision is the ratio of the number of samples actually belonging to the positive class and all the samples that were predicted to be of the positive class.
Precision= True Positives/(False Positives+True Positives)

- Recall is the ratio of the number of samples predicted correctly to be belonging to the positive class and all the samples that actually belong to the positive class.
Recall = True Positives/(False Negatives+True Positives)
- f1 Score is the harmonic mean of the recall and precision scores from the positive class. The f1-score takes into account both the precision and recall, and a high f1-score (closer to 1) implies that both the precision and recall are high.

```python
from sklearn.metrics import classification_report
cr = classification_report(y_test, np.argmax(test_pred, axis=1))
print(cr)
```

## ROC Curve

I will be creating an ROC curve to analyze the data.

The Receiver Operating Characteristic (ROC) curve is used for evaluating the performance of a classification model.

The plot shows the relationship between the true positive rate (TPR) and the false positive rate (FPR).

- A high TPR indicates that the model is good at identifying true positives.

- A low FPR indicates that the model is good at not falsely labeling negatives as positives.

The ROC curve is plotted with TPR on the y-axis and FPR on the x-axis. An ideal ROC curve is one that hugs the top-left corner of the plot, because this would indicate high TPR and low FPR.

Along with the ROC curve, the area under the ROC curve (AUC) is a  metric used to determine the model's performance. AUC outputs a single numerical value, which represents the model's ability to discriminate between positive and negative samples. The AUC is measured between 0 and 1. A higher AUC (closer to 1) indicates better performance.

The ROC is typically used for evaluating the performance of a binary classification model. However, in this analysis it was used to evaluate the performance of a multiclass classification model.

```python
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
import matplotlib.pyplot as plt

# Binarizing the labels
y_true_binarized = label_binarize(y_test, classes=[0, 1, 2, 3, 4])

# Computing ROC curve and ROC area for each class
fpr = dict() #false positive rate
tpr = dict() #true positive rate
roc_auc = dict() #auc
n_classes = y_true_binarized.shape[1] #number of classes

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_true_binarized[:, i], test_pred[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Plotting ROC curve for each class
plt.figure()

for i in range(n_classes):
    plt.plot(fpr[i], tpr[i], label=f'ROC curve (class {i}), AUC = {roc_auc[i]:.2f}')

plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc='lower right')
plt.show()

# Code adapted from: https://www.statology.org/plot-multiple-roc-curves-python/;
# https://stackoverflow.com/a/70002535;
# https://www.geeksforgeeks.org/multiclass-receiver-operating-characteristic-roc-in-scikit-learn/
```
